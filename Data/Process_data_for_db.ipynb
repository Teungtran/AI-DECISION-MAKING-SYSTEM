```python
import pandas as pd
import numpy as np
from collections import Counter
```

```python
df = pd.read_csv("ecommerce_customer_data_custom_ratios.csv")
```

## Prepare Data for table customers

```python
df_customer = df.copy()
df_customer
```

Cols: index_id,Customer_ID, Customer_Name, Age, Gender, Favoured_Payment_method, Favourdured_Product_Categort

```python
df_customer = df_customer.drop(columns=['Customer Age','Returns','Churn','Total Purchase Amount','Purchase Date'])
df_customer
```

```python
def most_common(lst): # function to count the most common element in a list
    counts = Counter(lst)
    if not counts:
        return None 
    return counts.most_common(1)[0][0]
```

```python
df_customer = df_customer.groupby("Customer ID", as_index=False, sort=False).agg(
    fullname = ("Customer Name", "first"),  
    Age = ("Age", "first"),
    Gender = ("Gender", "first"),
    Favoured_Payment_method = ("Payment Method", most_common),
    Favourdured_Product_Categort = ("Product Category", most_common))
df_customer.drop_duplicates(subset=['fullname'], keep='first', inplace=True)
```

```python
df_customer
```

```python
df_customer.columns.to_list()
```

```python
df_customer['index_id'] = None
df_customer = df_customer[['index_id','Customer ID',
 'fullname',
 'Age',
 'Gender',
 'Favoured_Payment_method',
 'Favourdured_Product_Categort']]
```

```python
df_customer
df_customer['Customer ID'] = 'KH'+ df_customer['Customer ID'].astype(str)
df_customer.rename(columns={"Customer ID": "customer_id"}, inplace=True)
```

```python
df_customer
```

```python
df_customer.isnull().sum()
```

```python
df_customer.to_csv("Customer_Infomation_Final.csv", index=False)
```

## Prepare Data for table orders

```python
df_order = df.copy()
```

```python
df_order
```

cols: Order_ID, Customer ID,	Purchase Date,Product Category,	Product Price,	Quantity,TotalPrice

```python
df_order['TotalPrice'] = df_order['Product Price']*df_order["Quantity"]
df_order = df_order.drop(columns=['Customer Age','Returns',
                                'Churn','Total Purchase Amount',
                                'Age','Gender'])
```

```python
df_order
```

```python
df_order.columns.to_list()
```

```python
df_order["OrderID"] = None
df_order = df_order[[
'OrderID',
'Customer ID',
 'Purchase Date',
 'Product Category',
 'Product Price',
 'Quantity',
 'Payment Method',
 'TotalPrice']]
```

```python
df_order.rename(
    columns={
        "OrderID": "OrderID",
        "Customer ID": "customer_id",
        "Purchase Date": "PurchaseDate",
        "Product Category": "ProductCategory",
        "Product Price": "ProductPrice",
        "Quantity": "Quantity",
        "Payment Method": "PaymentMethod",
        "TotalPrice": "TotalPrice",
    },
    inplace=True,
)
```

```python
df_order.info()
```

```python
df_order["ProductPrice"] = df_order["ProductPrice"].astype(float)
df_order["Quantity"] = df_order["Quantity"].astype(int)
df_order["TotalPrice"] = df_order["TotalPrice"].astype(float)
df_order.info()
```

```python
df_order['PurchaseDate'] = pd.to_datetime(df_order['PurchaseDate'], errors='coerce') 
df_order = df_order.dropna(subset=['PurchaseDate'])  
df_order['PurchaseDate'] = df_order['PurchaseDate'].dt.date  
```

```python
df_order['customer_id'] = "KH"+df_order["customer_id"].astype(str)
```

```python
df_order
```

```python
df_order.to_csv("Order_Infomation_Final.csv", index=False)
```
